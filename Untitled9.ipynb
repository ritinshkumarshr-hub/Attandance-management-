```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
n = int(input("Enter number of students: "))

students = []
for i in range(n):
    name = input(f"Enter name of student {i+1}: ")
    students.append(name)

students = np.array(students)

date = input("Enter date (DD-MM-YYYY): ")

attendance = []
for s in students:
    status = input(f"Mark attendance for {s} (P/A): ").upper()
    attendance.append(status)

df = pd.DataFrame({
    "Name": students,
    date: attendance
})

print("\nAttendance Record:")
print(df)

# Calculate attendance counts
attendance_counts = df[date].value_counts()
present_count = attendance_counts.get('P', 0)
absent_count = attendance_counts.get('A', 0)

print("\n" + "="*50)
print("ATTENDANCE SUMMARY")
print("="*50)
print(f"Total Students: {n}")
print(f"Present: {present_count}")
print(f"Absent: {absent_count}")


# Assuming 'date' variable holds the column name for tthe attendance data
# If not, please replace `date` with the actual column name, e.g., df.columns[1]

plt.figure(figsize=(8, 6))


sizes = [present_count, absent_count]
labels = ['Present', 'Absent']
colors = ['lightgreen', 'lightcoral']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

plt.title(f'Attendance on {date}\nTotal Students: {n}', fontsize=16)


plt.show()
```
